```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
boston_dataset = load_boston()
boston= pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston['Price'] = boston_dataset.target
boston.head(10)
```

```python
# correlation matrix -> relationship among features -> determine which features mostly affect the target
correlation_matrix = boston.corr().round(2)
# annot = True to print the values inside the square
sns.set(rc={'figure.figsize':(12,8)})
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
X = boston['RM']
Y = boston['Price']
plt.figure(figsize=(10, 5))
plt.scatter(X, Y, marker='o')
plt.title('Relationship between RM and Housing Price')
plt.xlabel('RM')
plt.ylabel('Price')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
X_train = np.c_[np.ones(X_train.size),X_train]
theta = np.zeros(2)
iterations = 1000;
alpha = 0.01;
```

```python
def computeCost(X,y,theta):
    predictions = X.dot(theta);
    squareFunc = (predictions - y)**2
    J = 1/(2*(y.size)) * np.sum(squareFunc);
    return J
```

```python
J_history = np.zeros(iterations)
thetas = []
for iteration in range(iterations):    
    x = X_train[:,1] #get all values in column 2nd of X, means x1
    h = theta[0] + (theta[1] * x)
    
    theta_zero = theta[0] - alpha*(1/Y_train.size) * np.sum(h-Y_train)
    theta_one = theta[1] - alpha*(1/Y_train.size) * np.sum((h-Y_train)*(x))
    
    theta = [theta_zero, theta_one]
    thetas.append(theta)
    #Save the cost J in every iteration    
    J_history[iteration] = computeCost(X_train, Y_train, theta)
    
theta_min = thetas[np.argmin(J_history)]
J_min = np.amin(J_history)
print(theta_min)
print(J_min)
```

```python
Y_pred_built = theta_min[0] + theta_min[1] * X_test
plt.figure(figsize=(7, 7))
plt.scatter(Y_test, Y_pred_built, c = 'red') 
plt.xlabel("Real result") 
plt.ylabel("Predicted result with self-built model") 
plt.axis([10,50,10,50])
plt.title("True value vs predicted value : Linear Regression") 
plt.show()
```
